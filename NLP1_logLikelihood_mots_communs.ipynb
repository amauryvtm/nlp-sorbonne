```python
import re as r
import glob
from math import log

entree_text1 = input("chemin d'accès n1 (sans guillemets) [comprenant le nom du fichier sous-jacent] du html :")
entree_text2 = input("chemin d'accès n2 (sans guillemets) [comprenant le nom du fichier sous-jacent] du html :")
texte1 = glob.glob(entree_text1)
texte2 = glob.glob(entree_text2)

dico1 = {}
for fichier1 in texte1:
    f = open(fichier1, "r", encoding="CP850")
    texte1 = f.read()
    f.close()

dico2 = {}
for fichier2 in texte2:
    f = open(fichier2, "r", encoding="CP850")
    texte2 = f.read()
    f.close()
    
    mots1 = r.findall("[a-zA-ZÀ-ÿ][a-zA-ZÀ-ÿ-]*", texte1)
    mots2 = r.findall("[a-zA-ZÀ-ÿ][a-zA-ZÀ-ÿ-]*", texte2)
    
    for mot1 in mots1:       
        if mot1 not in dico1:
            dico1[mot1] = 0
        dico1[mot1] = dico1[mot1] + 1

    for mot2 in mots2:       
        if mot2 not in dico2:
            dico2[mot2] = 0
        dico2[mot2] = dico2[mot2] + 1

#print(f"Mots différents : {len(dico1)}")

#transformer le dictionnaire en liste 
liste1 = []
for mot1 in dico1.items():
    liste1.append((mot1))

liste2 = []
for mot2 in dico2.items():
    liste2.append((mot2))

# Afficher mots et valeur donnée en dictionnaire
#print("fréquence des mots :")
for i in range(len(liste1)):
    mot1, freq1 = liste1[i]
    #print(f"{i+1} : {mot} ({freq})")

for i in range(len(liste2)):
    mot2, freq2 = liste2[i]

liste1.sort(key=lambda x: x[1], reverse=True)
liste2.sort(key=lambda x: x[1], reverse=True)

#for mot, freq in liste:
#    if freq>5:
#        print(mot,freq)

#print("liste 1 ; les 20 mots plus fréquents sont ")
for i in range(len(liste1)):
    mot1, freq1 = liste1[i]
    #print(f"{i+1}. {mot1} ({freq1})")

#print("liste 2 ; les 20 mots plus fréquents sont ")
for i in range(len(liste2)):
    mot2, freq2 = liste2[i]
    #print(f"{i+1}. {mot2} ({freq2})")

N1=sum(dico1.values())
N2=sum(dico2.values())


liste_commun = []
for mot1, freq1 in liste1:
    if mot1 in dico2:
        freq2 = dico2[mot1]
        E1 = N1*(freq1+freq2)/(N1+N2)
        E2 = N2*(freq1+freq2)/(N1+N2)
        LL = round( 2* (freq1*log(freq1/E1) + freq2*log(freq2/E2)), 5)
        liste_commun.append((LL, mot1))

liste_commun_sorted = sorted(liste_commun, reverse=True) 


print("liste log.Likelyhood ", liste_commun_sorted)






#python tab_freq.py
```
